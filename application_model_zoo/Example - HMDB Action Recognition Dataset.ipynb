```python
import cv2
import numpy as np
def overlay_text_on_video(input_video_path, output_video_path, class_name, score):
    cap = cv2.VideoCapture(input_video_path)

    if (cap.isOpened() == False): 
        print("Unable to read camera feed")

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))

    out = cv2.VideoWriter(output_video_path, 
                          cv2.VideoWriter_fourcc('M','J','P','G'), 
                          30, 
                          (frame_width,frame_height))

    while(True):
        ret, frame = cap.read()
        
        if ret == True: 
            cv2.putText(frame, 
                        class_name + ": " + str(score), 
                        (10, frame_height - 50), cv2.FONT_HERSHEY_SIMPLEX, 
                        1, 
                        (0, 0, 255),
                        3)
            out.write(frame)
        else:
            break  

    cap.release()
    out.release()

    cv2.destroyAllWindows() 
```

```python
import os

classes_list = sorted(os.listdir("rar_files/"));

f = open("classes.txt", 'w')
for i in range(len(classes_list)):
    f.write(classes_list[i] + "\n");
f.close();
```

```python
len(classes_list)
```

```python
import os
from tqdm import tqdm

complete_list = [];
for i in tqdm(range(len(classes_list))):
    folder_name = classes_list[i];
    vid_label = str(classes_list.index(folder_name));
    vid_list = os.listdir("rar_files/" + folder_name);
    for j in range(len(vid_list)):
        vid_name = vid_list[j];
        if(";" in vid_name):
            continue;
        vid_name_new = vid_name.replace("(", "\(")
        vid_name_new = vid_name_new.replace(")", "\)")
        vid_name_new = vid_name_new.replace(")", "\)")
        vid_name_new = vid_name_new.replace(";", "\;")
        vid_name_new = vid_name_new.replace("&", "\&")
        wr = str(vid_name) + " " + vid_label;
        os.system("cp rar_files/" + folder_name + "/" + vid_name_new + " videos/")
        complete_list.append(wr);
```

```python
complete_list[:5], len(complete_list), len(os.listdir("videos"))
```

```python
import random

random.shuffle(complete_list)
```

```python
complete_list[:5], len(complete_list)
```

```python
f = open("video_list.txt", 'w')
for i in range(len(complete_list)):
    f.write(complete_list[i] + "\n");
f.close();
```

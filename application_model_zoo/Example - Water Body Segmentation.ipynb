```python
import os
import cv2
import numpy as np 
from tqdm import tqdm
```

```python
img_list = os.listdir("dataset/Images");
```

```python
for i in tqdm(range(len(img_list))):
    img = cv2.imread("dataset/Images/" + img_list[i]);
    img = cv2.resize(img, (384, 384))
    cv2.imwrite("dataset/images_new/" + str(i) + ".png", img);

    mask = cv2.imread("dataset/Masks/" + img_list[i], 0);
    mask = cv2.resize(mask, (384, 384))
    mask[mask > 0] = 1;
    cv2.imwrite("dataset/masks_new/" + str(i) + ".png", mask)
```

## Training 

```python
import os
import sys
sys.path.append("Monk_Object_Detection/9_segmentation_models/lib/");
```

```python
from train_segmentation import Segmenter
```

```python
gtf = Segmenter();
```

```python
train_img_dir = "dataset/images_new";
train_mask_dir = "dataset/masks_new";

val_img_dir = "dataset/images_new";
val_mask_dir = "dataset/masks_new";
```

```python
classes_dict = {
    'background': 0, 
    'water': 1
};
classes_to_train = ['background', 'water'];
```

```python
gtf.Train_Dataset(train_img_dir, train_mask_dir, classes_dict, classes_to_train)
```

```python
gtf.Val_Dataset(val_img_dir, val_mask_dir)
```

```python
gtf.List_Backbones();
```

```python
gtf.Data_Params(batch_size=2, backbone="efficientnetb3", image_shape=[384, 384])
```

```python
gtf.List_Models();
```

```python
gtf.Model_Params(model="Unet")
```

```python
gtf.Train_Params(lr=0.0001)
```

```python
gtf.Setup();
```

```python
gtf.Train(num_epochs=300);
```

```python
gtf.Visualize_Training_History();
```
